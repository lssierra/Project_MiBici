```python
#write individually/in bulk? df to csv? 
input_dir = 'data'
output_file = 'data/full_data.csv'

n_files = 133
if os.path.exists(output_file):
    os.remove(output_file)
for i in range(1, n_files + 1):
    fn = os.path.join(input_dir, f'partial_data_{i}')
    # optionally: specify usecols and dtype to reduce memory
    # df = pd.read_csv(fn, usecols=[...], dtype={...})
    df = pd.read_csv(fn)
    df.to_csv(output_file, mode='a', header=not os.path.exists(output_file), index=False)
    del df
    gc.collect()

```
